## Exercise 4: Structured and Unstructured Data

In this exercise, we go over unstructured data, and why it's so important. Typically, such data is stored in JSON - a short form of Javascript Object Notation. JSON can be super structured (if we force it to be), like a table - or it can be flexible, like a list.


JSON could be used to store data about individuals like a contact card, or about sales, or about anything we desire it to. 

In this situation, we will create a contact card with the entry "Bob Martinez"

Such JSON typically follows the format below:






```python
import json

ContactCard = """{
  "firstName": "Bob",
  "lastName": "Martinez",
  "age": 40,
  "isStudent": false,
  "address": {
    "street": "123 Main St",
    "city": "Anytown"
  },
  "phoneNumbers": [
    {
      "type": "home",
      "number": "555-1234"
    },
    {
      "type": "work",
      "number": "555-5678"
    }
  ],
  "previousJobs": [
    {
      "type": "full-time",
      "title": "Engineer"
    },
    {
      "type": "full-time",
      "title": "Developer"
    }
  ],
  "manager": null
}"""
```

Notice how each piece of the JSON has a key, and then a value attached to it. This helps if we want to remember something about Bob Martinez - or if we need to call him to ask for his help fixing something!

We can use the Python JSON library to lookup those keys.

```python
data = json.loads(ContactCard)
```

```python
# Loop through the list of phone numbers to find the one with type "work"
work_phone = None
for phone in data['phoneNumbers']:
    if phone['type'] == 'work':
        work_phone = phone['number']
        break  # Stop looping once we find it

if work_phone:
    print(f"{data['firstName']}'s work phone is: {work_phone}")
else:
    print(f"{data['firstName']} does not have a work phone listed.")
```

Since JSON is unstructured, we had to loop through multiple "phone numbers" just to find out if Bob had one active. Unlike a SQL table, which can require phone number to not be null, JSON enforces no such restrictions. Bob could have had 6 work phone numbers, or just one, or none. So we have to loop and process the data ourselves to work with it! That makes things a little harder sometimes, but it also gives us immense flexibility.
