## Exercise 4: Structured and Unstructured Data - Solutions

```python
import json

PhoneBook = """{
  "firstName": "Bob",
  "lastName": "The Builder",
  "age": 40,
  "isStudent": false,
  "address": {
    "street": "123 Main St",
    "city": "Anytown"
  },
  "phoneNumbers": [
    {
      "type": "home",
      "number": "555-1234"
    },
    {
      "type": "work",
      "number": "555-5678"
    }
  ],
  "previousJobs": [
    {
      "type": "full-time",
      "title": "Engineer"
    },
    {
      "type": "full-time",
      "title": "Developer"
    }
  ],
  "manager": null
}"""
```

```python
data = json.loads(PhoneBook)
```

#### Task 1: Can you complete the code below to lookup Bob's previousJobs if they were full time?

```python
# 1. Initialize as an empty LIST to collect the job titles
previousJobs = [] 

# 2. Loop variable 'job_dict' holds each step
for job_dict in data['previousJobs']:
    # 3. Check the 'job_dict' variable
    if job_dict.get('type') == 'full-time': # The solution is 'full-time'
        # 4. Append the title you found to your list
        previousJobs.append(job_dict['title'])

# An empty list ([]) is 'False', a list with items is 'True', so no == statement is needed.
if previousJobs:
    print(f"Bob's full-time previous jobs are: {previousJobs}")
else: # if the list is not 'True' (i.e., empty)
    print("Bob was found, but he has no 'full-time' previous jobs listed.")
```

```python
data['services'] = ['Consulting', 'Development', 'Support']
updated_json_string = json.dumps(data, indent=2)
print(updated_json_string)
```
